```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Paraboloide hiperbolico

# Fórmula: z = (x^2 / a^2) - (y^2 / b^2)

# parametros
a = 1
b = 1

# Crear un rango de valores para x e y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calcular z para el paraboloide hiperbolico
Z = (X**2 / a**2) - (Y**2 / b**2)

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Configurar los ejes
ax.set_title('Paraboloide Hiperbólico')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# elipsoide

# Volumen de un elipsoide: V = (4/3) * π * a * b * c
# donde a = semieje mayor, b = semieje medio, c = semieje menor


# parametros
a = 3  # Radio en el eje X
b = 3  # Radio en el eje Y
c = 3  # Radio en el eje Z

# Crear una malla de valores para los angulos theta y phi
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Ecuaciones parametricas del elipsoide
X = a * np.sin(phi) * np.cos(theta)
Y = b * np.sin(phi) * np.sin(theta)
Z = c * np.cos(phi)

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie del elipsoide
ax.plot_surface(X, Y, Z, cmap='viridis')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# cono

# Volumen de un cono: V = (1/3) * π * r^2 * h
# donde r = radio de la base, h = altura


# parametros
a = 20
b = 20
c = 1

# Crear una malla de valores para x y y
x = np.linspace(-a, a, 100)
y = np.linspace(-b, b, 100)
X, Y = np.meshgrid(x, y)

# Definir la ecuacion del cono
Z = c * np.sqrt((X/a)**2 + (Y/b)**2)

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie del cono (parte superior e inferior)
ax.plot_surface(X, Y, Z, cmap='inferno')
ax.plot_surface(X, Y, -Z, cmap='inferno')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# cilindro eliptico

# Volumen de un cilindro eliptico: V = π * a * b * h
# donde a = semieje mayor, b = semieje menor, h = altura


# parametros
a = 2  # Radio eliptico en el eje x
b = 6  # Radio eliptico en el eje z

# Crear un rango de valores para u y v
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-5, 5, 100)

# Crear una malla de coordenadas para u y v
U, V = np.meshgrid(u, v)

# Define las ecuaciones parametricas
x = a * np.cos(U)
y = a * np.sin(U)
z = b * V

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie parametrica
ax.plot_surface(x, y, z, cmap='viridis')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# esfera

# Ecuación paramétrica de la esfera:
# \(x = r \sin(\phi) \cos(\theta)\)
# \(y = r \sin(\phi) \sin(\theta)\)
# \(z = r \cos(\phi)\)
# Donde \(r\) es el radio de la esfera.


# Radio de la esfera
r = 3

# Crear una malla de valores para los angulos theta y phi
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Ecuaciones parametricas de la esfera
X = r * np.sin(phi) * np.cos(theta)
Y = r * np.sin(phi) * np.sin(theta)
Z = r * np.cos(phi)

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie de la esfera
ax.plot_surface(X, Y, Z, cmap='coolwarm')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hiperboloide de una hoja

# Ecuacion del hiperboloide de una hoja:
# \(\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1\)

# Parametros - Cambia estos valores para ajustar la forma del hiperboloide
a = 3  # Parametro para el eje X
b = 2  # Parametro para el eje Y
c = 5  # Parametro para el eje Z

# Crear una malla de valores para u y v
u = np.linspace(0, 2 * np.pi, 100)  # Cambia el rango de u para ajustar la rotacion
v = np.linspace(-10, 10, 100)        # Cambia el rango de v para ajustar la altura
U, V = np.meshgrid(u, v)

# Ecuaciones parametricas del hiperboloide de una hoja
X = a * np.cosh(V) * np.cos(U)  # Cambia 'a' para modificar la amplitud en X
Y = b * np.cosh(V) * np.sin(U)  # Cambia 'b' para modificar la amplitud en Y
Z = c * np.sinh(V)              # Cambia 'c' para modificar la amplitud en Z

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie
ax.plot_surface(X, Y, Z, cmap='viridis')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# hiperboloide de dos hojas

# Ecuacion del hiperboloide de dos hojas:
# \(\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1\)


# Parametros
a = 3   # Parametro para el eje x
b = 2   # Parametro para el eje y
c = 5   # Parametro para el eje Z

# Crear una malla de valores para u y v
u = np.linspace(0, 2 * np.pi, 100)      # Cambia el rango de u para ajustar la rotacion
v = np.linspace(-2, 2, 100)             # Cambia el rango de v para ajustar la altura
U, V = np.meshgrid(u, v)

# Ecuaciones parametricas del hiperboloide de dos hojas
X = a * np.sqrt(1 + V**2) * np.cos(U)   #Cambia 'a' para modificar la amplitud en X
Y = b * np.sqrt(1 + V**2) * np.sin(U)   #Cambia 'b' para modificar la amplitud en y
Z = c * V                               # Cambia 'c' para modificar la amplitud en Z

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie
ax.plot_surface(X, Y, Z, cmap='plasma')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# paraboloide eliptico

# Parametros
a = 4  # Semieje en el eje X
b = 3  # Semieje en el eje Y
c = 1  # Factor de apertura

# Ecuacion del paraboloide eliptico:
# \(z = \left(\frac{x}{a}\right)^2 + \left(\frac{y}{b}\right)^2\)
# Donde \(a\) y \(b\) determinan la forma de la elipse en la base del paraboloide.

# Crear una malla de valores para x y y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Definir la ecuacion del paraboloide eliptico
Z = ((X/a)**2 + (Y/b)**2) * c

# Configurar la grafica 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar la superficie del paraboloide
ax.plot_surface(X, Y, Z, cmap='plasma')

# Configurar etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar la grafica
plt.show()
```

```python
#Funcion de dos variables

# Paso 1: Importar las librerias
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Paso 2: Definir las variables simbolicas y la nueva funcion
x, y = sp.symbols('x y')
f =  x**2 + y**2 # Definimos nuesta funcion

# Paso 3: Calcular las derivadas parciales simbolicas
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

# Imprimimos las derivadas parciales
print("Derivada parcial de f con respecto a x: df/dx =", df_dx)
print("Derivada parcial de f con respecto a y: df/dy =", df_dy)

# Paso 4: Convertir las funciones simbolicas a nuevas funciones numericas
f_num = sp.lambdify((x, y), f, "numpy")
df_dx_num = sp.lambdify((x, y), df_dx, "numpy")
df_dy_num = sp.lambdify((x, y), df_dy, "numpy")

# Paso 5: Crear un rango de valores para x e y
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Paso 6: Evaluar la funcion y las derivadas parciales en estos puntos
F_vals = f_num(X, Y)
df_dx_vals = df_dx_num(X, Y)
df_dy_vals = df_dy_num(X, Y)

# Paso 7: Graficar los resultados
fig = plt.figure(figsize=(12, 6))

# Gráfico de la funcion original f(x, y)
ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.plot_surface(X, Y, F_vals, cmap='viridis', edgecolor='none')
ax1.set_title("Función f(x, y) ")
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# Gráfico de la derivada parcial df/dx
ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax2.plot_surface(X, Y, df_dx_vals, cmap='coolwarm', edgecolor='none')
ax2.set_title("Derivada parcial df/dx")
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('df/dx')

# Gráfico de la derivada parcial df/dy
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.plot_surface(X, Y, df_dy_vals, cmap='plasma', edgecolor='none')
ax3.set_title("Derivada parcial df/dy")
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('df/dy')

# Ajustar el espaciado entre gráficos
plt.tight_layout()
plt.show()
```
